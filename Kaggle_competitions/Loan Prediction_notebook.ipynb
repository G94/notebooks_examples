```python
loan_train.Education.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

gender_mod=loan_train.Gender.mode()
loan_train["Gender"]=loan_train["Gender"].fillna(gender_mod[0])


married_mod=loan_train.Married.mode()
loan_train["Married"]=loan_train["Married"].fillna(married_mod[0])

dependents_mod=loan_train.Dependents.mode()
loan_train["Dependents"]=loan_train["Dependents"].fillna(dependents_mod[0])

self_mod=loan_train.Self_Employed.mode()
loan_train["Self_Employed"]=loan_train["Self_Employed"].fillna(self_mod[0])
```

```python
loan_train.count()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
gender = loan_train["Gender"]
gender_encoded = encoder.fit_transform(gender)
```

```python
married = loan_train["Married"]
married_encoded = encoder.fit_transform(married)
```

```python
education = loan_train["Education"]
education_encoded = encoder.fit_transform(education)
```

```python
dependents = loan_train["Dependents"]
dependents_encoded = encoder.fit_transform(dependents)
```

```python
self_education = loan_train["Self_Employed"]
self_education_encoded = encoder.fit_transform(self_education)
```

```python
property_area = loan_train["Property_Area"]
property_area_encoded = encoder.fit_transform(property_area)


loan_status = loan_train["Loan_Status"]
loan_status_encoded = encoder.fit_transform(loan_status)
```

```python
dtf_cat=pd.DataFrame({'gender':gender_encoded.tolist(), 'married':married_encoded.tolist(),'education':education_encoded.tolist(),'dependents':dependents_encoded.tolist(),'self_employed':self_education_encoded.tolist()})
dtf_cat.columns #OK
```

```python
dtf_property=pd.DataFrame({'property_area':property_area_encoded.tolist(),'Loan_Status':loan_status_encoded.tolist()})
dtf_property.reset_index(drop=True, inplace=True)
```

```python
loan_to_imput=loan_train[["LoanAmount","Loan_Amount_Term","Credit_History"]]
```

```python
from sklearn.preprocessing import Imputer
m_median = Imputer(strategy="median")
m_median.fit(loan_to_imput)
X = m_median.transform(loan_to_imput)
loan_imputed=pd.DataFrame(X,columns=loan_to_imput.columns)
loan_imputed.count()
```

```python
loan_appli=loan_train[["ApplicantIncome","CoapplicantIncome"]]
loan_appli.reset_index(drop=True, inplace=True)

l=pd.concat([dtf_cat,loan_appli], axis=1)

loan_concat2=pd.concat([l,loan_imputed], axis=1)

loan_concat2.reset_index(drop=True, inplace=True)

loan_train_final=pd.concat([loan_concat2,dtf_property], axis=1)
loan_train_final.reset_index(drop=True, inplace=True)
loan_train_final.columns


```
